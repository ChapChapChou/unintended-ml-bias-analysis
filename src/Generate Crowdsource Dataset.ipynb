```python
import pandas as pd
```

```python
debias = pd.read_csv('../data/toxicity_debiasing_data.tsv', sep = '\t')
```

```python
debias_random = pd.read_csv('../data/toxicity_debiasing_data_random.tsv', sep = '\t')
```

```python
debias.head()
```

```python
debias_random.head()
```

```python
mad_libs = pd.read_csv('../eval_datasets/bias_madlibs_89k.csv')
```

```python
mad_libs['dataset'] = 'madlibs'
```

```python
mad_libs.head()
```

```python
mad_libs.rename(columns={'Text': 'comment'}, inplace=True)
```

```python
mad_libs_good = mad_libs.query('Label == "NOT_BAD"')
```

```python
mad_libs_bad = mad_libs.query('Label == "BAD"')
```

```python
mad_libs_good['is_toxic'] = False
```

```python
mad_libs_bad['is_toxic'] = True
```

```python
mad_libs_bad.head()
```

```python
COLUMNS = ['comment', 'is_toxic', 'dataset']
mad_libs_good = mad_libs_good[COLUMNS]
mad_libs_bad = mad_libs_bad[COLUMNS]
```

```python
mad_libs_good = mad_libs_good.sample(n = 1000, random_state=12)
```

```python
mad_libs_bad = mad_libs_bad.sample(n = 1000, random_state=12)
```

```python
debias['dataset'] = 'debias'
debias_random['dataset'] = 'random'
```

```python
debias = debias.sample(n = 1000, random_state=12)
```

```python
debias_random = debias_random.sample(n = 1000, random_state=12)
```

```python
final_df = pd.concat([mad_libs_good, mad_libs_bad, debias, debias_random]).sample(frac = 1, random_state=12)
```

```python
final_df.head(20)
```

```python
final_df.query('dataset == "madlibs"').shape
```

```python
final_df.to_csv('../data/debias_and_madlibs_to_crowdsource.csv', index=False)
```
